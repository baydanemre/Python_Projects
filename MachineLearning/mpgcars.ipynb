```python
x = df_temp_2.loc[:,["mpg","displacement","horsepower","weight"]]
y = df_temp_2.loc[:,["origin"]]
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
x = imputer.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=40,random_state=42)
rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
rf.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test,y_pred)
acc
```

```python
import numpy as np

# y_train'i numpy dizisine dntrp ravel ile tek boyutlu hale getirme
y_train = np.array(y_train).ravel()

# Sonra kodunuza devam edebilirsiniz
accs = []
for i in range(1, 150):
    rf2 = RandomForestClassifier(n_estimators=i, random_state=42)
    rf2.fit(x_train, y_train)
    y_pred = rf2.predict(x_test)
    acc2 = accuracy_score(y_test, y_pred)
    accs.append(acc2)
```
